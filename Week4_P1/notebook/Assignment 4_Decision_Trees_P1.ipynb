```python
# Store the original number of rows
rows_before_cleaning = df.shape[0]

# Count duplicate rows before cleaning
duplicates_before_cleaning = df.duplicated().sum()

# Remove duplicate rows from the dataset
df_clean = df.drop_duplicates().copy()

# Convert currency columns from text to numeric values
currency_columns = ["Annual_Income", "Average_Order_Value"]

for column in currency_columns:
    df_clean[column] = df_clean[column].astype(str)
    df_clean[column] = df_clean[column].str.replace("$", "", regex=False)
    df_clean[column] = df_clean[column].str.replace(",", "", regex=False)
    df_clean[column] = pd.to_numeric(df_clean[column], errors="coerce")

# Print cleaning results
print("Rows before cleaning:", rows_before_cleaning)
print("Duplicate rows before cleaning:", duplicates_before_cleaning)
print("Rows after removing duplicates:", df_clean.shape[0])

# Display updated data types for the cleaned dataset
display(df_clean.dtypes)
```

```python
# Print a title for this output
print("Missing Values After Basic Cleaning")

# Print a separator line to make the output clear
print("-" * 50)

# Display missing values after removing duplicates and converting currency columns
display(df_clean.isnull().sum())
```

```python
# Store the target column name in a variable
target_column = "Converted"

# Print the target variable name
print("Target variable:")
print(target_column)
```

```python
# Print a title for the target values
print("Target Variable Values")

# Print a separator line
print("-" * 50)

# Display the unique values in the target column
print(df_clean[target_column].unique())

# Display the target class distribution
display(df_clean[target_column].value_counts())
```

```python
# Define columns that should not be used as predictive input features
columns_to_drop = ["Converted", "Customer_ID", "Campaign_Date"]

# Create X using all useful input features
X = df_clean.drop(columns=columns_to_drop)

# Create y using the target variable
y = df_clean[target_column]

# Print the shape of X and y
print("Feature matrix X shape:", X.shape)
print("Target vector y shape:", y.shape)

# Display feature columns
print("Input features:")
print(X.columns.tolist())
```

```python
# Identify categorical columns
categorical_features = X.select_dtypes(include=["object"]).columns.tolist()

# Identify numerical columns
numerical_features = X.select_dtypes(include=["int64", "float64"]).columns.tolist()

# Print categorical and numerical features
print("Categorical Features")
print("-" * 50)
print(categorical_features)

print("\nNumerical Features")
print("-" * 50)
print(numerical_features)
```

```python
# Import train_test_split to split the dataset
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

# Print the shape of each dataset
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
# Import preprocessing tools
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

# Create a preprocessing pipeline for numerical columns
numerical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median"))
])

# Create a preprocessing pipeline for categorical columns
categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
])

# Combine numerical and categorical preprocessing steps
preprocessor = ColumnTransformer(transformers=[
    ("num", numerical_transformer, numerical_features),
    ("cat", categorical_transformer, categorical_features)
])

print("Preprocessing pipeline created successfully.")
```

```python
# Import the Decision Tree classifier
from sklearn.tree import DecisionTreeClassifier

# Create the Decision Tree model inside a pipeline
# The pipeline first preprocesses the data and then trains the model
decision_tree_model = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("decision_tree", DecisionTreeClassifier(
        random_state=42,
        max_depth=3,
        min_samples_leaf=5
    ))
])

# Train the Decision Tree model using the training data
decision_tree_model.fit(X_train, y_train)

# Print a message after training is complete
print("Decision Tree model trained successfully!")
```

```python
# Import evaluation metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

# Make predictions on the training data
y_train_pred = decision_tree_model.predict(X_train)

# Make predictions on the testing data
y_test_pred = decision_tree_model.predict(X_test)

# Calculate training accuracy
training_accuracy = accuracy_score(y_train, y_train_pred)

# Calculate testing accuracy
testing_accuracy = accuracy_score(y_test, y_test_pred)

# Calculate evaluation metrics on the testing data
accuracy = accuracy_score(y_test, y_test_pred)
precision = precision_score(y_test, y_test_pred, pos_label="Yes")
recall = recall_score(y_test, y_test_pred, pos_label="Yes")
f1 = f1_score(y_test, y_test_pred, pos_label="Yes")

# Create a DataFrame to display the model results clearly
metrics_df = pd.DataFrame({
    "Metric": ["Training Accuracy", "Testing Accuracy", "Accuracy", "Precision", "Recall", "F1-score"],
    "Score": [training_accuracy, testing_accuracy, accuracy, precision, recall, f1]
})

# Display the model results
print("Decision Tree Evaluation Metrics")
print("-" * 50)
display(metrics_df)

# Print the classification report
print("Classification Report")
print("-" * 50)
print(classification_report(y_test, y_test_pred))
```

```python
# Import confusion matrix tools
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Create the confusion matrix
cm = confusion_matrix(y_test, y_test_pred, labels=["No", "Yes"])

# Display the confusion matrix as a table
confusion_matrix_df = pd.DataFrame(
    cm,
    index=["Actual No", "Actual Yes"],
    columns=["Predicted No", "Predicted Yes"]
)

print("Confusion Matrix")
print("-" * 50)
display(confusion_matrix_df)

# Plot the confusion matrix
ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["No", "Yes"]).plot()
plt.title("Decision Tree Confusion Matrix")
plt.savefig("confusion_matrix.png", bbox_inches="tight")
plt.show()
```

```python
# Print training and testing accuracy
print("Training Accuracy:", round(training_accuracy, 4))
print("Testing Accuracy:", round(testing_accuracy, 4))
print("Difference:", round(training_accuracy - testing_accuracy, 4))

# Calculate the difference between training and testing accuracy for each model
if training_accuracy - testing_accuracy > 0.10:
    print("The model may show signs of overfitting because training accuracy is much higher than testing accuracy.")
else:
    print("The model does not show strong signs of overfitting because training and testing accuracy are reasonably close.")
```

```python
# Get the trained preprocessing object from the pipeline
trained_preprocessor = decision_tree_model.named_steps["preprocessor"]

# Get the trained one-hot encoder from the categorical preprocessing pipeline
trained_encoder = trained_preprocessor.named_transformers_["cat"].named_steps["onehot"]

# Get feature names after one-hot encoding
encoded_categorical_features = trained_encoder.get_feature_names_out(categorical_features).tolist()

# Combine numerical and encoded categorical feature names
all_feature_names = numerical_features + encoded_categorical_features

# Get feature importance values from the trained Decision Tree model
feature_importance_values = decision_tree_model.named_steps["decision_tree"].feature_importances_

# Create a DataFrame of feature importances
feature_importance_df = pd.DataFrame({
    "Feature": all_feature_names,
    "Importance": feature_importance_values
}).sort_values(by="Importance", ascending=False)

# Display the top 10 most important features
print("Top 10 Important Features")
print("-" * 50)
display(feature_importance_df.head(10))
```

```python
# Plot the top 10 feature importances
top_features = feature_importance_df.head(10).sort_values(by="Importance")

plt.figure(figsize=(10, 6))
plt.barh(top_features["Feature"], top_features["Importance"])
plt.xlabel("Importance")
plt.ylabel("Feature")
plt.title("Top 10 Decision Tree Feature Importances")
plt.tight_layout()
plt.savefig("feature_importance.png", bbox_inches="tight")
plt.show()
```

```python
# Import tree plotting function
from sklearn.tree import plot_tree

# Get the trained Decision Tree object
trained_tree = decision_tree_model.named_steps["decision_tree"]

# Plot the Decision Tree
plt.figure(figsize=(22, 10))
plot_tree(
    trained_tree,
    feature_names=all_feature_names,
    class_names=trained_tree.classes_,
    rounded=True,
    fontsize=8
)
plt.title("Decision Tree Visualization")
plt.savefig("decision_tree_visualization.png", bbox_inches="tight")
plt.show()
```
